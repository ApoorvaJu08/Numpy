# Matplotlib.pyplot.plt()
Single Line Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
x = [1, 4, 5, 2, 3, 6]
plt.plot(x)
plt.show()
```

```python
import numpy as np
x = np.arange(10)
plt.plot(x)
plt.show()
```

```python
plt.plot(x, [y**2 for y in x])
plt.show()
```

# Multiline plots

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(10)
plt.plot(x, x**2)
plt.plot(x, x**3)
plt.show()
```

```python
plt.plot(x, x**2, x, x**3, x, x*2, x, 2**x)
plt.show()
```

```python
x = np.array([[1, 2, 6], [4, 5, 3]])
plt.plot(x)
plt.show()
```

```python
data = np.random.randn(2, 10)
print(data)
plt.plot(data[0], data[1])
plt.show()
```

# Grid Axes Labels

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(3)
plt.plot(x, x**2, x, x**3, x, x*2, x, 2**x)
plt.grid(True)
print(plt.axis())
plt.show()
```

```python
x = np.arange(3)
plt.plot(x, x**2, x, x**3, x, x*2, x, 2**x)
plt.grid(True)
#to explicitly set the axis
plt.axis([0, 2, 0, 8])
plt.show()
```

```python
x = np.arange(3)
plt.plot(x, x**2, x, x**3, x, x*2, x, 2**x)
plt.legend(['x**2', 'x**3', 'x*2', '2**x'], loc = 'upper center')
plt.grid(True)
print(plt.axis())
#to define separately the axis
plt.xlabel('x = np.arange(3)')
plt.xlim([0, 2])
plt.ylabel('y = f(x)')
plt.ylim([0, 8])
plt.title('Simple Plot Demo')
plt.show()
```

```python
x = np.arange(3)
plt.plot(x, x**2, label = 'x**2')
plt.plot(x, x**3, label = 'x**3')
plt.plot(x, x*2, label = 'x*2')
plt.plot(x, 2**x, label = '2**x')
plt.legend()
plt.grid(True)
print(plt.axis())
#to define separately the axis
plt.xlabel('x = np.arange(3)')
plt.xlim([0, 2])
plt.ylabel('y = f(x)')
plt.ylim([0, 8])
plt.title('Simple Plot Demo')
plt.savefig('test.png')
plt.show()
```

# Color line markers

```python
x = np.arange(5)
y = x

plt.plot(x, y+1, 'g')
plt.plot(x, y+0.5, 'y')
plt.plot(x, y, 'r')
plt.plot(x, y - 0.2, 'c')
plt.plot(x, y - 0.4, 'k')
plt.plot(x, y - 0.6, 'm')
plt.plot(x, y - 0.8, 'w')
plt.plot(x, y - 1, 'b')
plt.show()
```

```python
plt.plot(x, y+1, 'g', x, y+0.5, 'y', x, y, 'r', x, y - 0.2, 'c', x, y - 0.4, 'k', x, y - 0.6, 'm', x, y - 0.8, 'w', x, y - 1, 'b')
plt.show()
```

```python
plt.plot(x, y, '-', x, y+1, '--', x, y+2, '-.', x, y+3, ':')
plt.show()
```

```python
plt.plot(x, y, 'mo--') #combining color, marker and line style
plt.plot(x, y+1, 'g*-.')
plt.show()
```

```python
plt.plot(x, y, color = 'g', linestyle = '--', linewidth = 1.5, marker = '^', markerfacecolor = 'b', markeredgecolor = 'k', markeredgewidth = 1.5, markersize = 5)
plt.grid(True)
plt.show()
```

```python
x = y = np.arange(10)
plt.plot(x, y, 'o--')
plt.xticks(range(len(x)), ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
plt.yticks(range(1, 10, 1))
plt.show()
```
